Goal:

understand phenotype relationships

identify similar drugs

visualize nearest neighbors

Load embeddings

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

embeddings = np.load("data/processed/ssl_embeddings.npy")
labels = np.load("data/processed/ssl_labels.npy")
```

Compute similarity matrix

```python
sim_matrix = cosine_similarity(embeddings)
```

Find most similar compounds

```python
def get_neighbors(idx, top_k=5):

    sims = sim_matrix[idx]

    order = np.argsort(-sims)

    neighbors = order[1:top_k+1]

    return neighbors
```

Example analysis

```python
sample_idx = 0

neighbors = get_neighbors(sample_idx)

print("Query MOA:", labels[sample_idx])

for n in neighbors:
    print("Neighbor MOA:", labels[n])
```

Similarity heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.heatmap(sim_matrix[:50,:50], cmap="viridis")

plt.title("Embedding similarity")
plt.show()
```

What good results look like
