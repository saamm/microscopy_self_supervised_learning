```python
import zipfile
import os

# Specify the path to your zip file and the directory for extraction
zip_file_path = 'BBBC021_v1_images_Week1_22141.zip' # file path
extract_dir = 'Week1_22141' #  destination path

# Create the destination directory if it doesn't exist
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

# Open the zip file in read mode and extract all contents
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir) # Extract all files to the specified directory

print(f"Extracted all contents of '{zip_file_path}' to '{extract_dir}' directory.")
```
